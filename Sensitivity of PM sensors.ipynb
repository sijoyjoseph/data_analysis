This assignment deals with the mesuring the sensitivity of PM sensor data by changing the radius value for both PM10 and PM2.5

```python
from __future__ import division
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math

fields = ['time', 'pm10', 'pm25', 'rh']
df = pd.read_csv('..//datasets//rawdata.csv', usecols=fields)
df.head()
```

```python
# convert timestamp
df['time'] = pd.to_datetime(df['time'], unit='s')
df = df.set_index('time')
df = df.tz_localize('UTC').tz_convert('Asia/Kolkata')
df.head()
```

```python
"""
r_PM10 = 2.6, r_PM2.5 = 0.44 
radius_um = radius * (10**-6)  # Radius in micrometer
    vol = (4/3) * math.pi * (radius_um ** 3)
    density = 1.65 * (10**12)
    mass = density * vol
    K = 3531.5
    # Bring the count down to count/0.01ft3
    return round((count/100) * K * mass, 2)
"""

def compute(x, radius):
    radius_um = radius * (10**-6)
    vol = (4/3) * math.pi * (radius_um ** 3)
    density = 1.65 * (10**12)
    mass = density * vol
    K = 3531.5
    return round((x/100) * K * mass, 2)
    
r_PM10 = [2.0, 2.1, 2.2, 2.3, 2.4, 2.5, 2.6, 2.7, 2.8, 2.9, 3.0]

for val in r_PM10:
    column = 'PM10(r={0})'.format(str(val))
    df[column] = df['pm10'].apply(compute, args=(val,))
    
df.head()
```

```python
r_PM25 = [0.40, 0.41, 0.42, 0.43, 0.44, 0.45, 0.46, 0.47, 0.48, 0.49, 0.50]

for val in r_PM25:
    column = 'PM2.5(r={0})'.format(str(val))
    df[column] = df['pm25'].apply(compute, args=(val,))
    
df.head()
```

```python
def correct(x):
    ret_val = 0
    if (x['rh'] >= 0 and x['rh'] <= 39):
        ret_val = x['PM10(r=2.6)'] * x['rh'] * 13
    if (x['rh'] >= 40 and x['rh'] <= 49):
        ret_val = x['PM10(r=2.6)'] * x['rh'] * 8
    return ret_val

df['PM10_corrected(r=2.6)'] = df.apply(correct, axis=1)
df.head(100)
```
